### Date parser function
this function is used to change if the date column is in string then the data changes to date type in specific format ("%Y-%m-%d") using dateutil.parser

```
def date_parser(date_column):
  list1 = []
  for date in date_column:
    if type(date) == str:
      list1.append(dateutil.parser.parse(date).strftime("%Y-%m-%d"))
  return list1

```

## passing date columns to date_parser to get the similer format

```
df['Date2'] = date_parser(df['Date2'])
df['Date2']
```

## function to convert object type to datetime64[ns]
Creating a function to handle the date columns with irrespictive of format.
For Instance, In the date_format() function i have used exception handleing to specific date column to change the data type of different formts date string to datetime64[ns] type

```
def date_format(df_column):
  list1 = []
  for date in df_column:
    try:
      list1.append(datetime.datetime.strptime(date, "%m-%d-%Y"))
    except ValueError:
      list1.append(datetime.datetime.strptime(date, "%Y-%m-%d"))
  return list1    

```

```
df['Date2'] = date_format(df['Date2'])
```

## Funtion to drop the colmns have Pearson correlation more than 0.85

The Pearson's correlation coefficient is calculated as the covariance of the two variables divided by the product of the standard deviation of each data sample. The pearsonr() SciPy function can be used to calculate the Pearson's correlation coefficient between two data samples with the same length.

```
def drop_prsn_corel(df):
  pearsoncorr=  df.corr(method='pearson')
  df= df.drop(pearsoncorr>0.85,axis=1)
  return df
```
